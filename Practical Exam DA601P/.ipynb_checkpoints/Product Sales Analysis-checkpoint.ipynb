```python
import numpy as np
import pandas as pd 

import seaborn as sns
import matplotlib.pyplot as plt 

from sklearn.impute import KNNImputer 

sns.set(style='whitegrid')
```

```python
product_sales = pd.read_csv('product_sales.csv')
```

```python
product_sales.info()
```

```python
product_sales.describe()
```

```python
product_sales.head()
```

### Data Cleaning for 'sales_method'

```python
product_sales.sales_method.value_counts(dropna=False)
```

```python
# Convert strings to title case for sales_method
product_sales['sales_method'] = product_sales.sales_method.str.title() 
product_sales.sales_method.value_counts(dropna=False)
```

```python
# Replace words spelled incorrectly for sales_method
product_sales['sales_method'] = product_sales.sales_method.replace({'Em + Call': 'Email + Call'})
product_sales.sales_method.value_counts(dropna=False)
```

### Data Cleaning for 'revenue'

```python
# Implement KNN Imputation to fill in missing values
imputer = KNNImputer(n_neighbors=5)
product_sales['revenue']=imputer.fit_transform(product_sales[['revenue']])
```

```python
sns.displot(product_sales.revenue, bins=20)
```

### Data Cleaning for 'years_as_customer'

```python
# Replace values greater than 40 with 40 (If current year is 2024, max. value can only be 40.)
product_sales.loc[product_sales.years_as_customer > 40, 'years_as_customer'] = 40
```

## Data Validation

Changed all the strings to title case by using `product_sales.sales_method.str.title()` and replaced the values that are spelled differently by using `product_sales.sales_method.replace({'Em + Call': 'Email + Call'})`.
- customer_id: 15,000 unique IDs without missing values, same as the description. No cleaning is needed.
- nb_sold: Numeric values without missing values, same as the description. No cleaning is needed.
- revenue: Numeric values with 1,074 missing values. Used KNN imputation to fill in the missing values.
- years_as_customer: Numeric values without missing values. Expected max. value of 40 if the current year is 2024. Replaced the values that are larger than 40 with 40.
- nb_site_visits: Numeric values without missing values, same as description. No cleaning is needed.
- state: 50 states with no missing values, same as description. No cleaning is needed. 

## Exploratory Data Analysis 

```python
product_sales.head()
```

```python
product_sales.sales_method.value_counts(normalize=True)
```

```python
# How many customers were there for each approach? 

# Create the plot 
plt.figure(figsize=(12,6))
ax = sns.barplot(product_sales.sales_method.value_counts())

ax.set_xlabel('Sales Method', fontweight='bold')
ax.set_ylabel('Count / Number of Customers', fontweight='bold')
ax.set_title('Number of Customers by Sales Method', fontweight='bold')

plt.show()
```

From the last 6 weeks' record of a new line of office stationery launch, out of the 15,000 customers:
- 50% were approached by email
- 33% were approached by phone
- 17% were approached by both email and phone.

This indicates that the email approach was the most commonly used by the sales team, followed by the phone approach.

```python
# What does the spread of the revenue look like overall? And for each method?

# Create the plot 
plt.figure(figsize=(12,6))
ax = sns.histplot(data=product_sales, x='revenue', bins=15, hue='sales_method', kde=True)

# Set labels, title, and xticks 
ax.set_xlabel('Revenue from the sales', fontweight='bold')
ax.set_ylabel('Count / Number of Customers', fontweight='bold')
ax.set_title('Distribution of the Revenue', fontweight='bold')
ax.set_xticks([50, 75, 100, 125, 150, 175, 200, 225, 250])

plt.show()
```

There are high densities of customers who spent up to \\$140:
- Customers approached either by email or call mostly spend up to around \\$140.
- This indicates a high concentration of lower-spending customers in these two sales methods: Call or Email.
The lower density of higher-spending customers:
- A smaller number of customers approached by email and call spend between \\$150 and \\$240.
- This suggests that fewer customers approached by email and call fall into the higher spending range.

**Conclusion:**
The majority of customers approached by email or call tend to spend lower amounts (up to \\$140). There are fewer customers approached by email and call who spend higher amounts (\\$150 to \\$240). This indicates that the customers approached by email or call tend to spend less compared to those who spend more when they are approached by email and call.

```python
# Was there any difference in revenue over time for each of the methods? 

# Create the plot 
plt.figure(figsize=(12,6))
ax = sns.lineplot(data=product_sales, x='week', y='revenue', hue='sales_method')

# Set labels and title
ax.set_xlabel('Number of Weeks since Product Launch', fontweight='bold')
ax.set_ylabel('Revenue from the Sales', fontweight='bold')
ax.set_title('Revenue Over Time by Sales Method', fontweight='bold')

plt.show()
```

```python
product_sales.groupby(['week', 'sales_method'])['revenue'].mean().reset_index()
```

Since the start of the new product launch, there has been an increase in revenue across all three sales methods over the last 6 weeks: 
- The revenue from customers approached by call increased by \\$31.
- The revenue from customers approached by email increased by \\$40.
- The revenue from customers approached by both email and call \\$81.

This indicates that the combined approach of both email and call resulted in the highest increase in revenue.

```python
product_sales.head()
```

```python
# Find out if there are other differences between the customers in each group 

# Create the plot 
plt.figure(figsize=(12,6))
sns.boxplot(data=product_sales, x='sales_method', y='nb_sold')
plt.title('Distribution of Number of New Products Sold by Sales Method', fontweight='bold')
plt.xlabel('Sales Method', fontweight='bold')
plt.ylabel('Number of New Product Sold', fontweight='bold')
plt.show()
```

- 75% of the customers approached by both email and call bought 2 more items than 75% of the customers approached by either email or call alone.
- The median number of items bought by customers approached by email and call is 12 items, compared to a median of 10 items for customers approached by either email or call alone. 

This indicates that the customers approached by both email and call resulted bought 2 more items (at the 75th percentile) than those approached by either email or call alone. This combined approach resulted in a higher number of items sold.

```python
# Create the plot 
plt.figure(figsize=(12,6))
sns.boxplot(data=product_sales, x='sales_method', y='revenue')
plt.title('Distribution of Revenue by Sales Method', fontweight='bold')
plt.xlabel('Sales Method', fontweight='bold')
plt.ylabel('Revenue from the Sales', fontweight='bold')
plt.show()
```

Based on the boxplot:
- Customers approached by both call and email brought in the highest revenue.
- Customers approached by email alone generated the next highest revenue.
- Customers approached by call alone had the lowest revenue among the three groups.

There is no connection between the number of years as a customer and the sales method assigned to the customer.
